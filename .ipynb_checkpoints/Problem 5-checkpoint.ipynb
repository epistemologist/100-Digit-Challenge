# Problem 5
Let $ f(z) = 1 / \Gamma(z)$ where $ \Gamma(z) $ is the gamma function, and let $p(z)$ be the cubic polynomial that best approximates $f(z)$ on the unit disk in the supremum norm $\Vert{\cdot}\Vert_{\infty}$. What is $\Vert{f-p}\Vert_{\infty}$?

### Some mathematical preliminaries

Recall that the supremum norm of a function $f$ defined on domain $D$ is defined by $\Vert{f}\Vert_\infty := \max_{x \in D} |f(x)| $. 

If we let our optimal polynomial $p(x) := ax^3 + bx^2 + cx + d $, we have that $$ \epsilon := \Vert f-p \Vert_\infty = \max_{x \in D} |f(x)-p(x)| $$ 

where $D$ is the unit disk.

Now, note that because $1/\Gamma(x)$ and $p(x)$ are entire functions, we have that, by the [maximum principle](https://en.wikipedia.org/wiki/Maximum_modulus_principle) that the maximum lies on the boundary of $D$, i.e. the unit circle.

We therefore have 
$$\epsilon = \min_{a,b,c,d \in \mathbb{R}} \max_{\theta \in [0,2 \pi)} \left|\frac{1}{\Gamma(e^{i\theta})}- \left(a e^{3i\theta} + b e^{2i\theta}  + c e^{i \theta} + d \right) \right| $$

Firstly, we attempt out-of-the-box minimization routines. For an initial guess, we use the coefficients of the Maclaurin series of $f(z)$, namely:

$$ f(z) \approx f_0(z) = f(0) + f'(0) z + \frac{f''(0)}{2!} z^2  + \frac{f'''(0)}{3!} z^3$$

### Attempt 1: Brute-force using Scipy

````python
from math import cos, pi
from scipy.special import gamma
from scipy.optimize import minimize
from sympy import chebyshevt, lambdify
from sympy.abc import x
import numpy as np
from numpy import exp
import matplotlib.pyplot as plt

def f(x):
    return 1/gamma(x)

"""
Taken from "Numerical Recipes in C" 5.8 p193
"""
def chebfit(a,b,n,func):
    # Returns coeffs c[0..n-1] such that f(x) is about -c[0]/2 + sum([c[k]T_k(x) for k in range(n)]
    f = [None for i in range(n)]
    bma = 0.5*(b-a)
    bpa = 0.5*(b+a)
    for k in range(n):
        y = cos(pi*(k+0.5)/n)
        f[k] = func(y*bma+bpa)
    fac = 2./n
    c = [None for i in range(n)]
    for j in range(n):
        sum = 0
        for k in range(n):
            sum += f[k] * cos(pi*j*(k+0.5)/n)
        c[j] = fac*sum
    return c

c = chebfit(-1,1,4,f)
out = -c[0]/2 + sum([c[k]*chebyshevt(k,x) for k in range(len(c))])
f_hat_cheb = lambdify(x,out)

x = np.linspace(-1,1,100)
plt.plot(x,f(x))
plt.plot(x,f_hat_cheb(x))
plt.show()

def f_max(t,f_hat=f_hat):
    return abs(f(exp(1j*t))-f_hat(exp(1j*t)))

print("Minimum with Chebyshev interpolation: %s" % -minimize(lambda x: -f_max(x, f_hat_cheb), 0.5).fun)

from numpy import euler_gamma as y
f_hat_taylor = lambda z: z + y * z**2 + (y**2/2 - pi**2/12) * z**3

x = np.linspace(-1,1,100)
plt.plot(x,f(x))
plt.plot(x,f_hat_taylor(x))
plt.show()

print("Minimum with Taylor series interpolation: %s" % -minimize(lambda x: -f_max(x, f_hat_taylor), 0.5).fun)
````

# Attempt 2: Finding minimax polynomial with the Remez algorithm

Following the [Wikipedia article](https://en.wikipedia.org/wiki/Remez_algorithm)

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.misc import electrocardiogram
from scipy.optimize import minimize_scalar
from mpmath import *

poly = lambda arr: lambda x: sum([arr[i]*x**i for i in range(len(arr))])
l2 = lambda arr: sum([i*i for i in arr])
dist = lambda p,q: (p[0]-q[0])**2 + (p[1]-q[1])**2 
phi = (sqrt(5)+1)/2

def golden_section_search(f,a,b,tol=1e-12):
    # Taken from https://en.wikipedia.org/wiki/Golden-section_search
    # This is to *maximize* 
    c = b - (b-a) / phi
    d = a + (b-a) / phi
    while abs(c-d) > tol:
        if f(c) < f(d):
            b = d
        else:
            a = c
        c = b - (b-a)/phi
        d = a + (b-a)/phi
    return (a+b)/2

#print(remez(exp, 2, -1, 1))
mp.dps = 10
def remez_(f, n, a, b, tol=1e-6):
    # Solve for coefficients
    pts = [(x,f(x)) for x in linspace(a,b,n+2)]
    for i in range(100):
        X = [i[0] for i in pts]
        A = matrix(n+2,n+2)
        for i in range(n+2):
            for j in range(n+1):
                A[i,j] = X[i]**j
            A[i,n+1] = (-1)**(i+1)
        B = matrix([i[1] for i in pts])
        coeffs = lu_solve(A,B)[:-1]
        print(",".join([str(i) for i in coeffs]))
        P = poly(coeffs)
        # Exchange points (single point exchange)
        #error = lambda t: abs(P(t) - f(t))
        def error(t):
            #print(t)
            try:
                return float(abs(P(t)-f(t)))
            except:
                return 0
        plot(error, [a,b])
        max_x = X[np.argmax([error(i) for i in X])]
        eps = (b-a)/n
        max_x = minimize_scalar(error, method = "golden").x
        pts[np.argmin([dist(i,(max_x, f(max_x))) for i in pts])] = (max_x, f(max_x))


````
